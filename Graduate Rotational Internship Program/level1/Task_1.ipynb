```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
df
```

```python
plt.style.use('seaborn')
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(df['Hours'],df['Scores'],c='r');
ax.set(title='Marks vs Hours Studied', xlabel ='No of hours studied', ylabel ='Percentage of marks');
```

```python
df.plot(x ='Hours', y ='Scores', style = 'o')
plt.title('hours vs percentage')
plt.xlabel('hours studied')
plt.ylabel('percentage score')
plt.grid()
plt.show()
```

```python
x = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split  
x_train,x_test,y_train,y_test=train_test_split(x, y,random_state=0) 
```

```python
x_train.shape, y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression  
model=LinearRegression()  
model.fit(x_train, y_train) 

print("Training complete.")
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
plt.scatter(x_train,y_train,label='Scores')
plt.plot(x_train,model.predict(x_train),color='red',label='Traing data')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Line graph for Training data - Linear Regression')
plt.legend()
plt.show()
```

```python
df1=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df1
```

```python
sns.regplot(x='Actual',y='Predicted',data=df1)
```

```python
df1.plot(figsize=(10,5),kind='line')
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
